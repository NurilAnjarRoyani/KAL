# Perkalian Silang (Cross Product)

## Definisi Hasil Kali Silang

Hasil kali silang (cross product) adalah operasi biner antara dua vektor dalam ruang tiga dimensi (ℝ³) yang menghasilkan vektor baru. Vektor hasil ini:

1. Tegak lurus (orthogonal) terhadap kedua vektor asal.
2. Arahnya ditentukan oleh aturan tangan kanan (right-hand rule).
3. Besarnya (magnitudonya) sama dengan luas jajar genjang yang dibentuk oleh kedua vektor asal.



### Secara Matematis

Jika kita punya dua vektor:

$$
\mathbf{a} = \begin{bmatrix} a_1 \\ a_2 \\ a_3 \end{bmatrix}, \quad
\mathbf{b} = \begin{bmatrix} b_1 \\ b_2 \\ b_3 \end{bmatrix}
$$

Maka hasil kali silang (cross product) ditulis:

$$
\mathbf{a} \times \mathbf{b} = \begin{bmatrix}
a_2 b_3 - a_3 b_2 \\
a_3 b_1 - a_1 b_3 \\
a_1 b_2 - a_2 b_1
\end{bmatrix}
$$

Bisa juga ditulis dalam bentuk determinan matriks:

$$
\mathbf{a} \times \mathbf{b} =
\begin{vmatrix}
\mathbf{i} & \mathbf{j} & \mathbf{k} \\
a_1 & a_2 & a_3 \\
b_1 & b_2 & b_3
\end{vmatrix}
$$


## Contoh Menghitung Hasil Kali Silang:

Diberikan dua vektor 3 dimensi:

$$
\mathbf{a} = \begin{bmatrix} a_1 \\ a_2 \\ a_3 \end{bmatrix}, \quad
\mathbf{b} = \begin{bmatrix} b_1 \\ b_2 \\ b_3 \end{bmatrix}
$$

Maka:

$$
\mathbf{a} \times \mathbf{b} =
\begin{bmatrix}
a_2 b_3 - a_3 b_2 \\
a_3 b_1 - a_1 b_3 \\
a_1 b_2 - a_2 b_1
\end{bmatrix}
$$

Misalnya:

$$
\mathbf{a} = \begin{bmatrix} 3 \\ -1 \\ 2 \end{bmatrix}, \quad
\mathbf{b} = \begin{bmatrix} 1 \\ 4 \\ -2 \end{bmatrix}
$$

Hitung satu per satu komponennya:

1. Komponen x:

$$
a_2 b_3 - a_3 b_2 = (-1)(-2) - (2)(4) = 2 - 8 = -6
$$

2. Komponen y:

$$
a_3 b_1 - a_1 b_3 = (2)(1) - (3)(-2) = 2 + 6 = 8
$$

3. Komponen z:

$$
a_1 b_2 - a_2 b_1 = (3)(4) - (-1)(1) = 12 + 1 = 13
$$



### Hasilnya:

$$
\mathbf{a} \times \mathbf{b} = \begin{bmatrix} -6 \\ 8 \\ 13 \end{bmatrix}
$$

## Menggunakan Determinan Untuk Menemukan Perkalian Silang

### Contoh hasil kali silang determinan

Kita punya dua vektor:

$$
\mathbf{u} = \begin{bmatrix} 1 \\ 1 \\ 1 \end{bmatrix}, \quad
\mathbf{v} = \begin{bmatrix} 2 \\ 1 \\ 1 \end{bmatrix}
$$

Kita mau hitung:

$$
\mathbf{u} \times \mathbf{v}
$$



### Langkah 1: Gunakan rumus determinan:

$$
\mathbf{u} \times \mathbf{v} =
\begin{vmatrix}
\mathbf{i} & \mathbf{j} & \mathbf{k} \\
1 & 1 & 1 \\
2 & 1 & 1
\end{vmatrix}
$$



### Langkah 2: Ekspansi determinan:

$$
= \mathbf{i}(1 \cdot 1 - 1 \cdot 1)
- \mathbf{j}(1 \cdot 1 - 1 \cdot 2)
+ \mathbf{k}(1 \cdot 1 - 1 \cdot 2)
$$

$$
= \mathbf{i}(1 - 1)
- \mathbf{j}(1 - 2)
+ \mathbf{k}(1 - 2)
$$

$$
= \mathbf{i}(0)
- \mathbf{j}(-1)
+ \mathbf{k}(-1)
$$

$$
= 0\mathbf{i} + 1\mathbf{j} - 1\mathbf{k}
$$

### Hasilnya:

$$
\mathbf{u} \times \mathbf{v} = \begin{bmatrix} 0 \\ 1 \\ -1 \end{bmatrix}
$$

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Definisi vektor
u = np.array([1, 1, 1])
v = np.array([2, 1, 1])

# Hitung cross product
u_cross_v = np.cross(u, v)

# Tampilkan hasil
print("u × v =", u_cross_v)

# Visualisasi vektor
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Titik awal semua vektor dari (0,0,0)
origin = np.array([0, 0, 0])

# Gambar vektor u, v, dan u × v
ax.quiver(*origin, *u, color='blue', label='u')
ax.quiver(*origin, *v, color='green', label='v')
ax.quiver(*origin, *u_cross_v, color='red', label='u × v')

# Atur tampilan
ax.set_xlim([0, 2])
ax.set_ylim([0, 2])
ax.set_zlim([-2, 2])
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.legend()
ax.set_title('Visualisasi Cross Product')

plt.show()
```



## Sifat-Sifat Perkalian Silang (Cross Product)

Misalkan \$\mathbf{a}\$ dan \$\mathbf{b}\$ adalah dua vektor dalam ruang tiga dimensi (\$\mathbb{R}^3\$). Maka berlaku sifat-sifat berikut:

### 1. Sifat Anti-komutatif

$$
\mathbf{a} \times \mathbf{b} = - (\mathbf{b} \times \mathbf{a})
$$

> Maksudnya, jika urutan vektor dalam cross product ditukar, maka arah vektor hasilnya akan berbalik (negatif dari yang semula).



### 2. Tegak Lurus terhadap Kedua Vektor

$$
\mathbf{a} \cdot (\mathbf{a} \times \mathbf{b}) = 0, \quad \mathbf{b} \cdot (\mathbf{a} \times \mathbf{b}) = 0
$$

> Vektor hasil cross product selalu membentuk sudut 90° (orthogonal) terhadap vektor \$\mathbf{a}\$ dan \$\mathbf{b}\$.


### 3. Bersifat Distributif terhadap Penjumlahan

$$
\mathbf{a} \times (\mathbf{b} + \mathbf{c}) = \mathbf{a} \times \mathbf{b} + \mathbf{a} \times \mathbf{c}
$$

> Cross product mengikuti aturan distributif seperti pada operasi aljabar biasa.


### 4. Dapat Dikalikan dengan Skalar

$$
(k\mathbf{a}) \times \mathbf{b} = k (\mathbf{a} \times \mathbf{b}) = \mathbf{a} \times (k\mathbf{b})
$$

> Jika salah satu vektor dikalikan dengan bilangan skalar, maka skalar tersebut bisa dikeluarkan dari operasi cross product.


### 5. Cross Product Vektor dengan Dirinya Sendiri Bernilai Nol

$$
\mathbf{a} \times \mathbf{a} = \begin{bmatrix} 0 \\ 0 \\ 0 \end{bmatrix}
$$

> Karena dua vektor yang sejajar tidak membentuk bidang, maka luas jajar genjang yang terbentuk adalah nol, sehingga hasil perkalian silangnya adalah vektor nol.



## Sudut Antara Dua Vektor

Sudut antara dua vektor adalah sudut terkecil yang terbentuk ketika kedua vektor diletakkan dengan pangkal yang sama.

### Rumus Umum:

Untuk mencari sudut antara dua vektor, digunakan rumus:

$$
\cos\theta = \frac{\vec{A} \cdot \vec{B}}{|\vec{A}| \cdot |\vec{B}|}
$$

Keterangan:

* \$\vec{A} \cdot \vec{B}\$ adalah **dot product** dari kedua vektor, dihitung dengan:

  $$
  \vec{A} \cdot \vec{B} = a_1b_1 + a_2b_2 + a_3b_3
  $$

* \$|\vec{A}|\$ dan \$|\vec{B}|\$ masing-masing adalah **panjang atau magnitudo** vektor A dan B.

Untuk mendapatkan nilai sudut \$\theta\$, gunakan fungsi invers kosinus:

$$
\theta = \cos^{-1}\left( \frac{\vec{A} \cdot \vec{B}}{|\vec{A}| \cdot |\vec{B}|} \right)
$$



### Hubungan antara Dot Product dan Cross Product dengan Sudut:

* **Dot product** berkaitan dengan nilai **cosinus** sudut:

  $$
  \vec{A} \cdot \vec{B} = |\vec{A}||\vec{B}|\cos\theta
  $$

* **Cross product** berkaitan dengan **sinus** sudut:

  $$
  |\vec{A} \times \vec{B}| = |\vec{A}||\vec{B}|\sin\theta
  $$



### Contoh Perhitungan

Misalkan diberikan:

$$
\vec{A} = \langle 2, 3, 4 \rangle \quad \text{dan} \quad \vec{B} = \langle 1, 0, -1 \rangle
$$

#### A. Mencari Hasil Kali Silang \$\vec{A} \times \vec{B}\$

Gunakan rumus cross product 3 dimensi:

$$
\vec{A} \times \vec{B} = \langle a_2b_3 - a_3b_2,\ a_3b_1 - a_1b_3,\ a_1b_2 - a_2b_1 \rangle
$$

Substitusikan nilai komponen:

* Komponen-x: \$3(-1) - 4(0) = -3\$
* Komponen-y: \$4(1) - 2(-1) = 4 + 2 = 6\$
* Komponen-z: \$2(0) - 3(1) = -3\$

Jadi, hasilnya:

$$
\vec{A} \times \vec{B} = \langle -3,\ 6,\ -3 \rangle
$$



#### B. Menentukan Sudut antara \$\vec{A}\$ dan \$\vec{B}\$

**1. Hitung dot product:**

$$
\vec{A} \cdot \vec{B} = (2)(1) + (3)(0) + (4)(-1) = 2 - 4 = -2
$$

**2. Hitung magnitudo vektor:**

* \$|\vec{A}| = \sqrt{2^2 + 3^2 + 4^2} = \sqrt{29}\$
* \$|\vec{B}| = \sqrt{1^2 + 0^2 + (-1)^2} = \sqrt{2}\$

**3. Substitusi ke dalam rumus:**

$$
\cos\theta = \frac{-2}{\sqrt{29} \cdot \sqrt{2}} = \frac{-2}{\sqrt{58}}
$$

Gunakan kalkulator untuk menemukan \$\theta\$:

$$
\theta = \cos^{-1}\left( \frac{-2}{\sqrt{58}} \right) \approx \cos^{-1}(-0.2626) \approx 105.2^\circ
$$



### Kesimpulan:

* **Hasil cross product:**

  $$
  \vec{A} \times \vec{B} = \langle -3,\ 6,\ -3 \rangle
  $$

* **Sudut antara \$\vec{A}\$ dan \$\vec{B}\$:**

  $$
  \theta \approx 105.2^\circ
  $$



```python
import numpy as np
from numpy.linalg import norm
from math import degrees, acos
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Definisi dua vektor
A = np.array([2, 3, 4])
B = np.array([1, 0, -1])

# 1. Hasil kali silang (cross product)
cross_product = np.cross(A, B)

# 2. Dot product
dot_product = np.dot(A, B)

# 3. Magnitudo
magnitude_A = norm(A)
magnitude_B = norm(B)

# 4. Sudut (dalam radian dan derajat)
cos_theta = dot_product / (magnitude_A * magnitude_B)
theta_radian = acos(cos_theta)
theta_degree = degrees(theta_radian)

# 5. Tampilkan hasil perhitungan
print("Vektor A:", A)
print("Vektor B:", B)
print("Cross Product (A × B):", cross_product)
print("Dot Product (A · B):", dot_product)
print("Magnitude A:", magnitude_A)
print("Magnitude B:", magnitude_B)
print("Cos(θ):", cos_theta)
print("Sudut antara A dan B (derajat):", theta_degree)

# 6. Visualisasi dalam grafik 3D
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Titik awal vektor
origin = np.array([0, 0, 0])

# Gambar vektor A
ax.quiver(*origin, *A, color='r', label='Vektor A', arrow_length_ratio=0.1)
# Gambar vektor B
ax.quiver(*origin, *B, color='b', label='Vektor B', arrow_length_ratio=0.1)
# Gambar hasil cross product
ax.quiver(*origin, *cross_product, color='g', label='A × B (Cross Product)', arrow_length_ratio=0.1)

# Atur tampilan grafik
ax.set_xlim([0, max(A[0], B[0], cross_product[0], 1)])
ax.set_ylim([0, max(A[1], B[1], cross_product[1], 1)])
ax.set_zlim([0, max(A[2], B[2], cross_product[2], 1)])

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title(f'Visualisasi Vektor A, B dan A×B\nSudut antara A dan B ≈ {theta_degree:.2f}°')
ax.legend()
plt.grid(True)
plt.show()
```

## Implementasi

### PERSEGI

**1. Menentukan Titik-Titik Persegi**

Misalkan kita memiliki persegi dengan titik-titik sebagai berikut:

* A(0, 0)
* B(1, 0)
* C(1, 1)
* D(0, 1)

Titik-titik ini akan membentuk sebuah persegi jika disambungkan dalam urutan:
`A → B → C → D → A` (arah berlawanan jarum jam).



**2. Membuat Vektor dari Setiap Sisi**

Vektor dapat diperoleh dari selisih dua titik.

Contoh, vektor dari titik A ke B:

$$
\vec{AB} = B - A = (1 - 0,\ 0 - 0) = (1, 0)
$$

Hitung juga vektor sisi lainnya:

* **AB** = (1, 0)
* **BC** = (0, 1)
* **CD** = (-1, 0)
* **DA** = (0, -1)



**3. Menghitung Cross Product Antar Sisi**

Gunakan rumus untuk cross product dua vektor 2D:

$$
\vec{u} \times \vec{v} = u_x \cdot v_y - u_y \cdot v_x
$$

**Contoh Perhitungan:**

* **AB × BC**
  $(1 \cdot 1) - (0 \cdot 0) = 1$

* **BC × CD**
  $(0 \cdot 0) - (1 \cdot -1) = 1$

* **CD × DA**
  $(-1 \cdot -1) - (0 \cdot 0) = 1$

* **DA × AB**
  $(0 \cdot 0) - (-1 \cdot 1) = 1$



**4. Makna dari Hasil Perhitungan**

* Semua hasil bernilai **1** (positif)
* Ini menunjukkan bahwa setiap pertemuan sisi membentuk **belokan ke kiri** (berlawanan arah jarum jam / counter-clockwise)
* Kondisi ini sesuai dengan bentuk persegi yang diurutkan dalam arah **ccw**

Jika hasilnya **negatif**, berarti sudut-sudut terbentuk dari urutan searah jarum jam (cw).
Jika hasilnya **nol**, menunjukkan sisi-sisi tersebut **sejajar** (dan mungkin tidak membentuk persegi).



### Kesimpulan

Cross product pada persegi berguna untuk:

* Memastikan **arah rotasi antar sisi**
* Mengecek apakah bentuk benar-benar **persegi** (sudut harus tegak lurus)
* Menentukan apakah urutan titik **ccw** atau **cw**

```python
# Titik-titik persegi
A = (0, 0)
B = (1, 0)
C = (1, 1)
D = (0, 1)

# Fungsi untuk membuat vektor dari dua titik
def make_vector(p1, p2):
    return (p2[0] - p1[0], p2[1] - p1[1])

# Fungsi untuk menghitung cross product 2D
def cross_product(v1, v2):
    return v1[0] * v2[1] - v1[1] * v2[0]

# Buat vektor sisi-sisi persegi
AB = make_vector(A, B)
BC = make_vector(B, C)
CD = make_vector(C, D)
DA = make_vector(D, A)

# Hitung cross product antar sisi
cp1 = cross_product(AB, BC)
cp2 = cross_product(BC, CD)
cp3 = cross_product(CD, DA)
cp4 = cross_product(DA, AB)

# Cetak hasil
print("Cross Product AB × BC:", cp1)
print("Cross Product BC × CD:", cp2)
print("Cross Product CD × DA:", cp3)
print("Cross Product DA × AB:", cp4)

# Interpretasi
crosses = [cp1, cp2, cp3, cp4]

if all(c > 0 for c in crosses):
    print("Semua sudut membentuk belokan ke kiri (counter-clockwise)")
elif all(c < 0 for c in crosses):
    print("Semua sudut membentuk belokan ke kanan (clockwise)")
elif any(c == 0 for c in crosses):
    print("Ada sudut yang sejajar (cross product = 0), kemungkinan bukan persegi")
else:
    print("Titik-titik tidak berurutan secara konsisten (mungkin bukan persegi)")
```
